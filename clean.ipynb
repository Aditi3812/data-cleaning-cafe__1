```python
import pandas as pd
```

```python
df = pd.read_csv("dirty_cafe_sales.csv")
df
```

1. item , Quantity, price col has nan values to be removed.
2. total Spent	Payment Method	Location nan to either mean / median.

```python
df["Transaction Date"].info()
```

```python
df.drop_duplicates()
```

```python
df["Item"]= df["Item"].fillna("")
df
```

```python
for i in list(df.columns):
    df[i]= df[i].fillna("")
df
```

```python
df['Total Spent']= df['Total Spent'].replace("ERROR","")
df['Payment Method']=df['Payment Method'].str.replace("UNKNOWN","")
df['Location']=df['Location'].str.replace("UNKNOWN","")
```

```python
df['Location']= df['Location'].str.replace('[^a-zA-Z0-9]',' ',regex=True)
df
```

```python
df['Transaction Date']=df['Transaction Date'].replace('[^0-9]',"/", regex= True)
df
```

```python
df['Quantity']=df['Quantity'].replace("", df['Quantity'].mean)
df['Price Per Unit']=df['Price Per Unit'].replace("", df['Price Per Unit'].mean)
df['Total Spent']=df['Total Spent'].replace("", df['Total Spent'].mean)
```

```python
df['Payment Method']=df['Payment Method'].replace("", df['Payment Method'].mean)
```

```python
for i in df.columns:
    df[i] = df[i].replace("",df[i].mean)
df
```
